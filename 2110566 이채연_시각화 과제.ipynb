#1 BMI와 허리둘레 및 HDL콜레스테롤 간 상관관계 분석

[데이터 분석 과정]
1. 키와 신장 데이터로 BMI를 계산해 BMI 구간을 나눠, 고도비만인이 가장 많은 지역을 선정한다.
2. 고도비만인이 가장 많은 지역에서 BMI와 허리둘레의 상관관계를 시각화한다.
3. 추가로 고도비만인이 가장 많은 지역에서 BMI와 HDL콜레스테롤의 상관관계를 시각화해, "'비만의 척도'인 체질량지수(BMI)가 높을수록 혈중 HDL콜레스테롤 농도가 낮다"라는 연구 결과를 직접 증명한다.
(실제 연구 결과 관련 기사 : https://health.chosun.com/site/data/html_dir/2021/02/02/2021020201068.html)

*사용 그래프 : 막대, 산점도

```python
#키를 m 단위로 변환한 열 추가하기

health['height']=health['신장(5Cm단위)']*0.01
display(health.head())
```

```python
#m 단위로 변환한 키와 체중을 이용해 'bmi' 계산해 새로운 열 추가하기

health['bmi']=health['체중(5Kg 단위)']/health['height']**2
display(health.head(20))
```

```python
#bmi 구간이라는 새로운 열 만들고 일단 0으로 채우기

health['bmi_구간']=0
display(health.head(20))
```

```python
#bmi 지수를 구간별로 나누어 새로운 열 추가하기

health.loc[health['bmi'] < 18.5, 'bmi_구간'] = '저체중'
health.loc[(health['bmi'] >= 18.5) & (health['bmi'] < 23), 'bmi_구간'] = '정상'
health.loc[(health['bmi'] >= 23) & (health['bmi'] < 25), 'bmi_구간'] = '과체중'
health.loc[(health['bmi'] >= 25) & (health['bmi'] < 30), 'bmi_구간'] = '경도비만'
health.loc[(health['bmi'] >= 30) & (health['bmi'] < 35), 'bmi_구간'] = '중증비만'
health.loc[health['bmi'] >= 35, 'bmi_구간'] = '고도비만'
display(health.head(20))
```

```python
#고도비만인 행만 추출해서 새로운 데이터 프레임(health_new) 만들기
health_new=health[health['bmi_구간'] == '고도비만']
display(health_new.head(10))
```

```python
#시도별 고도비만 수 구해서 막대 그래프 그리기
CountStatus = pd.value_counts(health_new['시도코드'].values, sort=True)
CountStatus.plot.bar()

###인사이트
#시각화 결과 고도비만인이 가장 많은 시도는 41, 즉 경기도다.
```

```python
#경기도 행만 추출해서 새로운 데이터 프레임(health_new2) 만들기
health_new2=health[health['시도코드'] == 41 ]
display(health_new2.head(10))
```

```python
#bmi와 허리둘레 간 상관관계 그래프 그려 인사이트 도출하기
sns.scatterplot(data = health_new2, x = 'bmi', y = '허리둘레',hue='bmi_구간')

###인사이트
#여러 지역 중 고도비만인이 가장 많은 경기도 데이터만 추출해서 경기도인들의 bmi와 허리둘레의 상관관계를 분석하니,
#bmi가 높아질수록 허리둘레도 높아지는 우상향 그래프가 그려졌다. 
```

```python
#추가로 bmi와 HDL콜레스테롤 간 상관관계 인사이트 증명하기
sns.regplot(x='bmi', y='HDL콜레스테롤', lowess=True, data=health_new2,
            line_kws={'color': 'red'}, scatter_kws={'color': 'grey'})

###인사이트
#또한 유용한 물질인 HDL콜레스테롤과 bmi 간 상관관계를 분석한 결과,
#bmi가 높을수록, 즉 비만일수록 HDL콜레스테롤 수치가 낮아짐을 확인할 수 있다.
#빨간 추세선을 통해 그 결과가 더 명확히 보여진다.
#"비만의 척도인 체질량지수(BMI)가 높을수록 혈중 HDL콜레스테롤 농도가 낮다"는 연구결과를 본 데이터로도 직접 확인할 수 있었다.
```

#2 음주 여부와 ALT, AST 수치 간 상관관계 분석

[데이터 분석 과정]
1. 음주를 더 많이 하는 성별을 도출한다.
2. 그 성별의 연령별 ALT 수치 평균을 구한다.
3. 그 성별의 연령별 AST 수치 평균을 구한다. 
4. 2번과 3번을 그래프로 그려 인사이트를 도출한다.

*사용 그래프 : 막대, 선

```python
#성별별 음주여부를 알 수 있는 교차표 생성
health2=pd.crosstab(health['성별코드'],health['음주여부'], margins=True)
display(health2.head())

###인사이트
#남성이 여성보다 음주를 하는 사람의 수가 많다.
```

```python
#위 교차표 시각화
health2.plot(kind='barh', stacked=True, title='성별별 음주여부')
```

```python
#성별이 남자인 행만 추출하기
health3=health[health['성별코드'] == 'M' ]
display(health3.head(10))
```

```python
#성별이 남자인 데이터 중에서 연령대와 ALT 열만 추출하기

health4=health3[['연령대코드(5세단위)','(혈청지오티)ALT']]
display(health4.head(10))
```

```python
#남성 연령 그룹별 ALT 평균구하기
health5=health4.groupby(['연령대코드(5세단위)'], as_index=False).mean()
health5
```

```python
#성별이 남자인 데이터 중에서 연령대와 AST 열만 추출하기

health6=health3[['연령대코드(5세단위)','(혈청지오티)AST']]
display(health6.head(10))
```

```python
#남성 연령 그룹별 AST 평균구하기
health7=health6.groupby(['연령대코드(5세단위)'], as_index=False).mean()
health7
```

```python
#연령과 ALT 및 AST 간 상관관계 시각화 후 인사이트 도출하기
sns.lineplot(data = health5, x = '연령대코드(5세단위)', y = '(혈청지오티)ALT')
sns.lineplot(data = health7, x = '연령대코드(5세단위)', y = '(혈청지오티)AST')

###인사이트
#ALT는 간기능을 나타내는 수치로 농도가 높을수록 간세포 손상 정도가 심한 것이다.
#파란색 선그래프(ALT)그래프가 우하향하는 것으로 보아 나이가 증가할수록 간세포 손상 정도가 약화된다는 것을 알 수 있다.
#특히 연령대코드 9 즉, 40~44세 남성들의 간세포 정도가 가장 심하다.
#그에 반해 AST는 ALT와는 달리 뚜렷한 상향이나 하향 변화를 보이지 않는다.(주황색 선 그래프)
#AST는 간 기능을 나타내는 혈액검사상의 수치, 간세포 이외에 심장, 신장, 뇌 , 근육 등에도 존재하는 효소인데,
#이러한 세포들이 손상을 받는 경우 농도가 증가한다.
#AST는 ALT와는 달리 연령에 따라 큰 변화가 없음을 알 수 있다.
```

#3 금연 캠페인 시범 지역과 대상 정하기

[데이터 분석 과정]
1. 흡연자가 가장 많은 지역을 도출한다.
2. 그 지역에서 연령별 흡련자 수를 구해, 흡연자가 가장 많은 연령대를 구한다.
3. 그 연령대에서 남여 수를 비교한다.
4. 만약 금연 캠페인을 한다면, 첫 시범으로 어디에 누구를 대상으로 하면 좋을지 인사이트를 도출한다. 

*사용 그래프 : 막대, 파이

```python
#지역별 음주여부를 알 수 있는 교차표 생성
health10=pd.crosstab(health['시도코드'],health['흡연상태'], margins=True)
health10

###인사이트
#흡연자가 가장 많은 지역은 경기도(41)이다.
```

```python
#경기도 행만 추출해서 새로운 데이터 프레임 만들기
health11=health[health['시도코드'] == 41 ]
display(health11.head()) 
```

```python
#경기도 내 연령별 흡연자 수 구하기
health12=pd.crosstab(health11['연령대코드(5세단위)'],health11['흡연상태'], margins=True)
health12

###인사이트
#경기도 내 흡연자 수가 가장 많은 연령은 9, 즉 40~44세이다.
```

```python
#경기도 40~44세에 해당하는 데이터만 추출해 새로운 데이터 프레임 생성하기
health13=health11[health11['연령대코드(5세단위)'] == 9]
display(health13.head(10))
```

```python
#경기도 40~44세 흡연자 성별에 따른 인원수 비교하기
health14=pd.crosstab(health13['성별코드'],health13['흡연상태'], margins=True)
health14

###인사이트
#경기도 40~44세 흡연자는 여자 13명, 남자 104명으로 남자가 더 많다.
```

```python
#위 교차표 시각화하기
health20=health14.drop('All', axis=0)
health21=health20.drop('All', axis=1) 
health21.plot(kind='barh', stacked=True, title='성별별 흡연여부')
```

```python
#위 표(health14) 중 흡연자 부분만 비율로 환산하기
13/117*100

###인사이트
#경기도 40~44세 흡연자 중 여성 비율은 11%이다.
```

```python
#위 표(health14) 중 흡연자 부분만 비율로 환산하기
104/117*100

###인사이트
#경기도 40~44세 흡연자 중 남성 비율은 89%이다.
```

```python
#위 두 남여 비율을 파이 그래프로 시각화하기
ratio = [11, 89]
labels = ['Female', 'Male']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()

###인사이트
#지역 중 흡연자 수가 가장 많은 곳은 경기도다.
#경기도민 중 흡연자 수가 가장 많은 연령은 40~44세이다.
#경기도 40~44세 흡연자 중 여자 13명, 남자 104명으로 남자가 더 많다.
#비율로 환산하면 여자가 11%, 남자가 89%이다.
#따라서 만약 금연 캠페인 시작 전 첫 시범 운영 지역을 고른다면,
#지역은 경기도, 대상은 40~44세 남성으로 하면 효과적일 것이다.
```
